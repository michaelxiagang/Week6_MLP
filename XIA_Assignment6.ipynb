```python
RANDOM_SEED = 1 # seed value for random number generator             
import numpy as np
import pandas as pd
import os
import sklearn
import tensorflow as tf
from datetime import datetime
from scipy import stats as st
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score # evaluation metrics
from sklearn.preprocessing import StandardScaler # used for variable scaling data
from sklearn.preprocessing import MinMaxScaler  # used for variable scaling data
os.getcwd() 
```

```python
# Read train data set
mydata_train=pd.read_csv('train.csv')
mydata_test =pd.read_csv('test.csv')
print(mydata_train.shape)
print(mydata_test.shape)
```

```python
# Display structure of train dataset
pd.set_option('display.max_colwidth', -1)
print("train data set:\n", mydata_train.head())
print("test data set:\n", mydata_test.head())
```

## Using sklearn neural network model to re-rerun assignment 5
## Two layer MLP classification

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(max_iter=100)

# set up response and independent variables 
X_train=mydata_train.iloc[:, 1:784].values
y_train=mydata_train.iloc[:,0].values
X_test =mydata_test.iloc[:, :783].values
```

```python
# Define a hyper-parameter space to search
parameter_space = {
    'hidden_layer_sizes': [(10,10), (20,20)],
    'activation': ['tanh', 'relu'],
    'solver': ['lbfgs', 'sgd', 'adam'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant','adaptive'],
}
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(mlp, parameter_space, n_jobs=-1, cv=3)

clf.fit(X_train,y_train)
```

```python
# Best paramete set
print('Best parameters found:\n', clf.best_params_)
```

```python
# Accuracy Score
print("Accuracy: ",clf.score(X_train,y_train))
```

```python
# All results
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
param = clf.cv_results_['params']
df = np.array([means,stds,param])
df = pd.DataFrame(df).T
df.columns = ['means','stds','param']
df.sort_values(["means"], axis=0,ascending=False, inplace=True)
df
```

```python
# Using best parameter to model the training data set
start=datetime.now()
mlp1 = MLPClassifier(activation='relu', alpha=0.05, hidden_layer_sizes=(20, 20), learning_rate='adaptive', solver='adam')
mlp1.fit(X_train,y_train)
end=datetime.now()
print("time:", end-start)  #get the run time
```

```python
print("Accuracy:", mlp1.score(X_train,y_train))
```

```python
# Run the prediction on X_test data set
y_pred=mlp1.predict(X_test)
y_pred1 = pd.DataFrame(y_pred)
y_pred1.index = y_pred1.index +1
y_pred1.to_csv("Xia_Assignment6_2_mlp_2layer_best.csv")
```
